### Subplot

```python
x
```

```python
y
```

```python
y2 = y*x
```

```python
plt.plot(y)
plt.plot(y2)

plt.show();
```

```python
plt.plot(x, y)
plt.plot(x, y2)

plt.show();
```

```python
# row, col, pos

plt.subplot(1, 2, 1)
plt.plot(x, y,
         marker = 'o',
         color = 'g')

plt.subplot(1, 2, 2)
plt.plot(x, y2,
         marker = '*',
         color = 'b',
         markersize = 10)

plt.show();
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.1, 0.6, 0.4, 0.3])

ax1.plot(x, y, 'r')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Y plot')

ax2.plot(x, y2, 'g')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Y2 Plot')

plt.show();
```

### xlim, ylim, xticks and yticks

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 4))

ax[0].plot(x, y, x, y2)

ax[1].plot(x, y**2, 'k')

ax[1].set_ylim([0, 5_000])
ax[1].set_xlim([-5, 5])

plt.show();
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 4))

ax[0].plot(x, y, x, y2)
ax[0].set_xticks([1, 4, 9])

ax[1].plot(x, y**2, 'k')
ax[1].set_xticks([1, 4, 9])
ax[1].set_xticklabels([r'$\alpha$', r'$\beta$', r'$\delta$'])

plt.show();
```
