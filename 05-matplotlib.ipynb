```python
y = y.cumsum()
```

```python
plt.plot(x, y);
```

```python
plt.plot(x, y, 
         color = 'g',
         marker = 'o',
         linestyle = '--');
```

### Label

```python
plt.plot(x, y, 
         color = 'r',
         marker = 'o',
         linestyle = '--');

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Demo plot')

plt.show();
```

### Scater, bar, hist and box plots

```python
x, y
```

```python
plt.scatter(x, y,
            linewidths=2,
            )
plt.show();
```

```python
b = [10,23, 43, 5, 66]
a = [ 'a', 'b', 'c', 'd', 'e']

plt.bar(a, b,
        width = 0.9)
plt.show();
```

```python
plt.hist(y, 
         rwidth = 0.8,
         bins = 30)
plt.show();
```
